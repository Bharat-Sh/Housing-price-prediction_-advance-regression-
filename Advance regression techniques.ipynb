```python
input_cols = ['MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'Street',
       'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig',
       'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType',
       'HouseStyle', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd',
       'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType',
       'MasVnrArea', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinSF1',
       'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', '1stFlrSF', '2ndFlrSF',
       'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath',
       'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual',
       'TotRmsAbvGrd', 'Functional', 'Fireplaces', 'FireplaceQu', 'GarageType',
       'GarageYrBlt', 'GarageFinish', 'GarageCars', 'GarageArea', 'GarageQual',
       'GarageCond', 'PavedDrive', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'PoolQC',
       'Fence', 'MiscFeature', 'MiscVal', 'MoSold', 'YrSold', 'SaleType',
       'SaleCondition']
```

```python
target_col = 'SalePrice'
```

```python

inputs_df = data_df[input_cols].copy()
target_df = data_df[target_col].copy()

```

```python
#getting the list of numeric columns
numeric_cols = inputs_df.select_dtypes(include=['int64', 'float64']).columns.tolist()

#getting the list of categorical columns
categorical_cols = inputs_df.select_dtypes('object').columns.tolist()
```

```python

data_df[numeric_cols].head()
```

Impute Numerical Data


```python
missing_counts = inputs_df[numeric_cols].isna().sum().sort_values(ascending=False)
missing_counts[missing_counts > 0]
```

```python

from sklearn.impute import SimpleImputer
# Impute missing values for numeric columns with the mean
imputer = SimpleImputer(strategy = 'median')
imputer.fit(data_df[numeric_cols])
inputs_df[numeric_cols] = imputer.transform(inputs_df[numeric_cols])
```

```python
missing_counts = inputs_df[numeric_cols].isna().sum().sort_values(ascending=False)
missing_counts[missing_counts > 0] 
```

Scaling of numerical data

```python
from sklearn.preprocessing import MinMaxScaler

# Create the scaler
scaler = MinMaxScaler()

# Fit the scaler to the numeric columns
scaler.fit(data_df[numeric_cols])
```

```python
# Transform and replace the numeric columns
inputs_df[numeric_cols] = scaler.transform(inputs_df[numeric_cols])
```

```python
inputs_df[numeric_cols].describe().loc[['min', 'max']]
```

                    Categorical Colums
    

```python
inputs_df[categorical_cols].nunique().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import OneHotEncoder

# 1. Create the encoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# 2. Fit the encoder to the categorical colums
encoder.fit(data_df[categorical_cols])

# 3. Generate column names for each category
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
print(len(encoded_cols))

# 4. Transform and add new one-hot category columns
inputs_df[encoded_cols] = encoder.transform(inputs_df[categorical_cols])
```

```python

inputs_df.head()
```

```python

inputs_df[numeric_cols + encoded_cols].head()
```

```python

log_targets = np.log(target_df + 1)
```

         Train test split
         

```python

from sklearn.model_selection import train_test_split
```

```python
train_inputs, val_inputs, train_targets, val_targets = train_test_split(inputs_df[numeric_cols + encoded_cols], 
                                                                        log_targets, 
                                                                        test_size=0.25, 
                                                                        random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

# Create and train the linear regression model
linreg = LinearRegression()
linreg.fit(train_inputs, train_targets)

# Predict on validation set
val_preds = linreg.predict(val_inputs)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(val_targets, val_preds)
rmse = np.sqrt(mse)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
# Predict on the training set
train_preds = linreg.predict(train_inputs)

# Calculate MSE for training set
train_mse = mean_squared_error(train_targets, train_preds)
train_rmse = np.sqrt(train_mse)

# Calculate MSE for validation set (already computed as mse, rmse)
print("Train MSE:", train_mse)
print("Train RMSE:", train_rmse)
print("Validation MSE:", mse)
print("Validation RMSE:", rmse)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Create and train the Gradient Boosting Regressor model
gbr = GradientBoostingRegressor(random_state=42)
gbr.fit(train_inputs, train_targets)

# Predict on validation set
val_preds_gbr = gbr.predict(val_inputs)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(val_targets, val_preds)
rmse = np.sqrt(mse)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
# Predict on the training set
train_preds = gbr.predict(train_inputs)

# Calculate MSE for training set
train_mse = mean_squared_error(train_targets, train_preds)
train_rmse = np.sqrt(train_mse)

# Calculate MSE for validation set (already computed as mse, rmse)
print("Train MSE:", train_mse)
print("Train RMSE:", train_rmse)
print("Validation MSE:", mse)
print("Validation RMSE:", rmse)
```

```python
from sklearn.linear_model import Ridge

# Create and train the Ridge regression model
ridge = Ridge(alpha=1.0, random_state=42)
ridge.fit(train_inputs, train_targets)

# Predict on validation set
val_preds_ridge = ridge.predict(val_inputs)
```

```python
# Predict on the training set
train_preds = ridge.predict(train_inputs)

# Calculate MSE for training set
train_mse = mean_squared_error(train_targets, train_preds)
train_rmse = np.sqrt(train_mse)

# Calculate MSE for validation set (already computed as mse, rmse)
print("Train MSE:", train_mse)
print("Train RMSE:", train_rmse)
print("Validation MSE:", mse)
print("Validation RMSE:", rmse)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate RMSE for Linear Regression
rmse_linreg = np.sqrt(mean_squared_error(val_targets, val_preds))

# Calculate RMSE for Gradient Boosting Regressor
rmse_gbr = np.sqrt(mean_squared_error(val_targets, val_preds_gbr))

# Calculate RMSE for Ridge Regression
rmse_ridge = np.sqrt(mean_squared_error(val_targets, val_preds_ridge))

print("Validation RMSE (Linear Regression):", rmse_linreg)
print("Validation RMSE (Gradient Boosting):", rmse_gbr)
print("Validation RMSE (Ridge Regression):", rmse_ridge)
```
